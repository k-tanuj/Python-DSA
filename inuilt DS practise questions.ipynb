## IN-BUILT DATA STRUCTURE : PRACTISE QUESTIONS

##### 1. SUM OF ALL THE ELEMENTS IN A LIST

```python
numbers = [10,20,30,40,50]
total = 0
for x in numbers:
    total += x
print(total)

#alternate
# print(sum(list))
```

##### 2. Largest Element in a List



```python
numbers = [10,20,30,40,50]
largest=0
for i in numbers:
    if i>largest:
        largest = i
print(largest)

#alternate
#maxx = max(list)
#print(maxx) 
```

##### 3. Remove Duplicate in a List

```python
numbers = [10,20,30,20,40,30,50]

numbers = list(set(numbers)) #set doesnt contains duplicates

print(numbers)
```

#### 4. Check if all elements in a list are Unique

```python
numbers = [10,20,30,20,40,30,50]
unique = True
for i in range(len(numbers)):
    for j in range(i+1,len(numbers)):
        if numbers[i]==numbers[j]:
            unique = False

print(unique)

```

#### 4. Program to Reverse a List

```python
numbers = [10,20,30,40,50]
reverse = list(reversed(numbers))

print(reverse)
```

#### 5. Count Number of Odd and Even Elements in a List

```python
numbers = [11,22,36,47,59]
even = 0
odd = 0
for i in numbers:
    if i%2==0:
        even+=1
    else:
        odd+=1

print("odd:",odd)
print("even:",even)
```

#### 6. Maximum difference between two consecutive elements in a list.

```python
numbers = [11,22,36,47,59]
diff = 0

for i in range(len(numbers)-1):
    if numbers[i+1] - numbers[i] > diff:
        diff = numbers[i+1] - numbers[i]
print(diff)
```

##### 7. merge to sortedlist

```python
numbers = [11,22,36,47,59]
numbers2 = [10,20,30,40,50]

numbers3 = numbers + numbers2

numbers3.sort()

print(numbers3)
```

#### 8. Merge 2 List into Dictionary

```python
alph = ['a','b','c','d','e']
numbers2 = [10,20,30,40,50]

result = {}

for i in range(len(alph)):
    result[alph[i]]=[numbers2[i]]

print(result)
```
